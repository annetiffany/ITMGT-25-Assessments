```python
import time

data_points = []

for i in range(100):
    print (i)
    iss = requests.get('http://api.open-notify.org/iss-now.json')
    iss_data = iss.json()
    data_points.append(iss_data)
    time.sleep(2)
```

```python
data_points
```

```python
header = ['timestamp','latitude','longitude']

body = []
for record in data_points:
    row = []
    row.append(record['timestamp'])
    row.append(record['iss_position']['latitude'])
    row.append(record['iss_position']['longitude'])
    body.append(row)
```

```python
import csv

with open('international_space_station.csv','w') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(body)
```
